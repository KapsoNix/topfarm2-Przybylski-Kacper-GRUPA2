```python
import numpy as np
import matplotlib.pyplot as plt

import topfarm  # test importu

x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])

plt.figure(figsize=(6, 4))
plt.scatter(x, y)
plt.title("Layout turbin – Zadanie 2.1")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True)
plt.axis("equal")
plt.show()
```

Oto wyjaśnienie tego kodu w 5 punktach:

Import bibliotek – Na początku importujemy narzędzia do obliczeń (numpy) i rysowania wykresów (matplotlib), a także testujemy dostęp do biblioteki topfarm.

Definiowanie danych – Tworzymy dwie tablice liczb: współrzędne X (0, 500, 1000, 1500) i współrzędne Y (0, 0, 500, 500), które reprezentują położenia turbozin wiatrowych.

Tworzenie wykresu – Otwarty zostaje nowe okno wykresu o rozmiarze 6×4 jednostek.

Rysowanie punktów i formatowanie – Na wykres nanoszone są cztery punkty, dodane zostają opisy (tytuł, nazwy osi w metrach), siatka i osie ustawione w równej skali.

Wyświetlenie wyniku – Komenda plt.show() wyświetla gotowy wykres na ekranie, pokazując wizualizację rozmieszczenia turbin.

```python
# Zadanie 2.1 – Wykres z kolorami i etykietami
x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])
colors = ['blue', 'red', 'green', 'yellow']
labels = ['T1', 'T2', 'T3', 'T4']

plt.figure(figsize=(8, 6))

# Rysowanie punktów z kolorami i etykietami
for i in range(len(x)):
    plt.scatter(x[i], y[i], c=colors[i], s=150, label=labels[i])
    plt.text(x[i]+30, y[i]+30, labels[i], fontsize=11, fontweight='bold')

plt.title("Layout turbin – Zadanie 2.1 z kolorami i etykietami")
plt.xlabel("x [m]")
plt.ylabel("y [m]")
plt.grid(True)
plt.axis("equal")
plt.legend()
plt.show()
```

Kod:

Definiuje cztery pozycje turbin (X i Y)
Przypisuje każdej turbinie unikalny kolor (niebieski, czerwony, zielony, żółty)
Rysuje każdy punkt za pomocą pętli z odpowiadającym mu kolorem
Dodaje etykiety dla każdej turbiny (T1, T2, T3, T4)
Ustawia odpowiednie opisy osi, tytuł, siatkę i równą skalę osi
Wyświetla legendę z kolorami

```python
# Zadanie 2.1 – Wykres z nowymi kolorami i etykietami
x = np.array([0, 500, 1000, 1500])
y = np.array([0, 0, 500, 500])
colors = ['black', 'brown', 'orange', 'tan']
labels = ['Turbina A', 'Turbina B', 'Turbina C', 'Turbina D']

plt.figure(figsize=(12, 10))

# Rysowanie punktów z kolorami i etykietami
for i in range(len(x)):
    plt.scatter(x[i], y[i], c=colors[i], s=300, label=labels[i], edgecolors='black', linewidth=2)
    plt.text(x[i]+50, y[i]+50, labels[i], fontsize=13, fontweight='bold')

plt.title("Layout turbin – Zadanie 2.1 z kolorami i etykietami", fontsize=16, fontweight='bold')
plt.xlabel("x [m]", fontsize=13)
plt.ylabel("y [m]", fontsize=13)
plt.grid(True, alpha=0.3)
plt.axis("equal")
plt.legend(fontsize=12, loc='upper left')
plt.tight_layout()
plt.show()
```
